**Class Assignment:** (Convert a user given name as in the shown format)
* **Input:** moHan dAs kaRaM chAnd gaNDhi
* **Output:** M D K C Gandhi

```python
# Input: moHan dAs kaRaM chAnd gaNDhi
# Output: M D K C Gandhi

def my_string_processing(mystr):
    flag = 1
    outputstr = ""
    for i in range(len(mystr)):
        ch = mystr[i]
        if (flag == 1):
            outputstr += ch.upper()
            position = i
            flag = 0
        if (ch == " "):
            flag = 1
            outputstr += " "
    outputstr += mystr[position + 1:].lower()
    return outputstr

name_str = "moHan dAs kaRaM chAnd gaNDhi"
result_str = my_string_processing(name_str)
print (f"Original name string is: {name_str}...")
print (f"Processed name string is: {result_str}")
print ("End of the program...")
```

>Take a sentence from the user as input. The sentence may contain upper and lower case characters. Find the number of vowels and consonants in the given sentence.

```python
mystr = input("Please enter your sentence: ")
vcount = ccount = 0
for ch in mystr.lower():
    if (ch.isalpha()):
        if (ch in "aeiou"): vcount += 1
        else: ccount += 1
print (f"So the sentence '{mystr}' contains number of vowels = {vcount} and consonants = {ccount}...")
print ("End of the program...")
```

> Take n number of integers from the user and put them in a list with out using any loop. And sort them and print them.
Now take one by one number from the newly created list and put them in two lists, even_list and odd_list
depending upon the fact whether the number is even or odd respectively. And print both two newly created lists.

```python
list1 = [int(x) for x in input("Please enter your set of integer numbers: ").split()]
print (f"Original list is {list1}...")
list1.sort()
print (f"Sorted list is {list1}...")
even_list = []
odd_list = []
for item in list1:
    if (item % 2 == 0): even_list.append(item)
    else: odd_list.append(item)
print (f"Even list is {even_list}...")
print (f"Odd list is {odd_list}...")
print ("End of the program...")
```

```python
def switch_case_demo(month_no):
    switcher = {1:"January",
                2:"February",
                3:"March",
                4:"April",
                5:"May",
                6:"June",
                7:"July",
                8:"August",
                9:"September",
               10:"October",
               11:"November",
               12:"December"}
    return switcher.get(month_no, "Invalid Month No.")

print (switch_case_demo(3))
print (switch_case_demo(11))
print (switch_case_demo(9))
print (switch_case_demo(30))
```

### Set Processing

```python
# creating a set from different data sources
# from a list
list1 = ["apple", "guava", False, 2.45, "grape", "banana", True, 100, "guava", False, 2.45, "grape"]
set1 = set(list1)
print (set1, len(set1), type(set1), id(set1))

# from a tuple
tuple1 = ("apple", "guava", False, 2.45, "grape", "banana", True, 100, "guava", False, 2.45, "grape")
set1 = set(tuple1)
print (set1, len(set1), type(set1), id(set1))

# from a set directly
set1 = {"apple", "guava", False, 2.45, "grape", "banana", True, 100, "guava", False, 2.45, "grape"}
print (set1, len(set1), type(set1), id(set1))
```

```python
# set operations: union, intersection, difference, symmetric difference
# union operation
snakes = {'viper', 'python', 'cobra'}
print (snakes)
languages = {'c++', 'kotlin', 'python'}
print (languages)
result = languages.union(snakes)
print (result)
result = languages | snakes
print (result)
print (languages)
print (snakes)
```

```python
# intersection operation
snakes = {'viper', 'python', 'cobra'}
print (snakes)
languages = {'c++', 'kotlin', 'python'}
print (languages)
result = languages.intersection(snakes)
print (result)
result = languages & snakes
print (result)
print (languages)
print (snakes)
languages.intersection_update(snakes)
print (languages)
print (snakes)
```

```python
# difference operation
snakes = {'viper', 'python', 'cobra'}
print (snakes)
languages = {'c++', 'kotlin', 'python'}
print (languages)
result = languages.difference(snakes)
print (result)
result = languages - snakes
print (result)
print (languages)
print (snakes)
languages.difference_update(snakes)
print (languages)
print (snakes)
```

```python
# symmetric difference operation
snakes = {'viper', 'python', 'cobra'}
print (snakes)
languages = {'c++', 'kotlin', 'python'}
print (languages)
result = languages.symmetric_difference(snakes)
print (result)
result = languages ^ snakes
print (result)
print (languages)
print (snakes)
languages.symmetric_difference_update(snakes)
print (languages)
print (snakes)
```

```python
set1 = set([100, 200, 300])
set2 = set([400, 500, 600, 700, 800])
set3 = set([400, 600, 700])
print (set1.issuperset(set2), set2.issuperset(set3))
print (set1.isdisjoint(set2), set2.isdisjoint(set3))
print (set1.issubset(set2), set2.issubset(set3), set3.issubset(set2))
```

### Frozen Set Processing
