> Rewrite the previous code in such a way that the program will keep on asking user to enter new pair of input numbers until it can perform successful division operation.

```python
while True:
    try:
        num1 = int(input("Please enter the first number (between -100 to +100 inclusive): "))
        num2 = int(input("Please enter the second number (between -100 to +100 inclusive): "))
        if (num1 < -100 or num2 < -100): raise NameError("Below-100")
        if (num1 > 100 or num2 > 100): raise NameError("Above+100")
        result = num1 / num2
        print (f"So {num1} / {num2} = {result}...")
    except ValueError as ve:
        print ("ValueError: Please provide integer numeric data only...!!!")
        print ("ValueError: Error Type:", type(ve))
        print ("ValueError: Error Message:", ve)
    except ZeroDivisionError as zde:
        print ("ZeroDivisionError: Division By ZERO is Illegal...!!!")
        print ("ZeroDivisionError: Error Type:", type(zde))
        print ("ZeroDivisionError: Error Message:", zde)
    except NameError as ne:
        print ("NameError: Input data is out of range...!!!")
        print ("NameError: Error Type:", type(ne))
        print ("NameError: Error Message:", ne)
        if (str(ne) == "Below-100"): print ("NameError: Input value is less than -100...")
        if (str(ne) == "Above+100"): print ("NameError: Input value is greater than +100...")
    except Exception as ex:
        print ("Exception: Some unknown exception has occurred...!!!")
        print ("Exception: Error Type:", type(ex))
        print ("Exception: Error Message:", ex)
    else:
        print ("Else: Executing the ELSE Block...")
        print ("Else: Had a smooth execution...")
        break
    finally:
        print ("Finally: Executing the FINALLY Block...")
        print ("Finally: This block executes always...")
print ("End of the program...")
```

```python
class Error(Exception):
    """Base class for other Exception classes..."""
class ValueTooSmallError(Error):
    """Raised when the input value is too small..."""
    def __init__(self, error_code, error_message):
        self.error_code = error_code
        self.error_message = error_message
class ValueTooLargeError(Error):
    """Raised when the input value is too large..."""
    def __init__(self, error_code, error_message):
        self.error_code = error_code
        self.error_message = error_message
# main code
number = 10
while True:
    try:
        input_num = int(input("Please enter your guess number: "))
        if (input_num < number): raise ValueTooSmallError("Error-001", "Too small value...")
        if (input_num > number): raise ValueTooLargeError("Error-002", "Too large value...")
        break
    except ValueTooSmallError as vtse:
        print ("Your input value is too small... Try Again...!!!")
        print ("Error Code:", vtse.error_code)
        print ("Error Message:", vtse.error_message)
        print ()
    except ValueTooLargeError as vtle:
        print ("Your input value is too large... Try Again...!!!")
        print ("Error Code:", vtle.error_code)
        print ("Error Message:", vtle.error_message)
        print ()
print ("Congratulations !!! Your guess is Perfectly CORRECT...!!!")
print ("End of the game...!!!")
```
